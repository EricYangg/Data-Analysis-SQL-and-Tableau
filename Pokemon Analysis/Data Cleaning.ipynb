Will replace with 'Unavailable' since it's only 2 rows 

```python
cleaned_df = df.copy()
cleaned_df.replace('—', 'Unavailable', inplace=True)
cleaned_df
```

Split Type into Primary Type and Secondary Type

```python
col_name = 'Type'
new_col1 = 'Primary Type'
new_col2 = 'Secondary Type'

#Get index of 'Type' column
replace_index = cleaned_df.columns.get_loc(col_name)

#Split the 'type' column into primary and secondary types
cleaned_df[[new_col1, new_col2]] = cleaned_df[col_name].str.split(' ', expand=True)

#Drop Type column
cleaned_df.drop(columns=[col_name], inplace=True)

#Reorder columns by inserting new columns in place of original Type column 
column_order = cleaned_df.columns.tolist()
column_order.insert(replace_index, new_col1)
column_order.insert(replace_index + 1, new_col2)

#Remove the columns at the end
cleaned_df = cleaned_df[column_order]
cleaned_df = cleaned_df.iloc[:, :-2]

cleaned_df
```
